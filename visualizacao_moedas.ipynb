# 💱 Visualização de Dados Cambiais
Este notebook carrega os arquivos `.parquet` gerados pelo pipeline e exibe gráficos de tendências para as moedas USDBRL, EURBRL e ARSBRL.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import datetime

# Data atual para localizar a pasta
today = datetime.now().strftime('%Y-%m-%d')
data_dir = f'../data/{today}/'

# Carregar todos os arquivos parquet
symbols = ['USDBRL', 'EURBRL', 'ARSBRL']
dfs = {}
for symbol in symbols:
    file_path = os.path.join(data_dir, f'{symbol}.parquet')
    if os.path.exists(file_path):
        dfs[symbol] = pd.read_parquet(file_path)

# Mostrar as primeiras linhas de cada DataFrame
for symbol, df in dfs.items():
    print(f"\n📈 {symbol}")
    display(df.head())
```

## 📊 Gráfico de Fechamento Diário

```python
plt.figure(figsize=(12, 6))

for symbol, df in dfs.items():
    plt.plot(df['timestamp'], df['close'], label=symbol)

plt.title('Valor de Fechamento Diário')
plt.xlabel('Data')
plt.ylabel('Cotação (close)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
